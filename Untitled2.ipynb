```python
cleantable['Pregnancies'] = cleantable['Pregnancies'].apply(lambda x: x if x<9 else 9)
```

```python
cleantable['Insulin'].isna().sum()
```

```python
insuline_agg_all = cleantable.groupby(['Pregnancies', 'Outcome']).agg({'Insulin': 'mean'})
insuline_agg_all
```

```python
cleantable.columns
```

```python
name = pd.merge(cleantable, insuline_agg_all, how='left', on=['Outcome','Pregnancies'])
name
```

```python
import numpy as np
name['Insuline'] = np.where(name['Insulin_x'].isna(), name['Insulin_y'].round(), name['Insulin_x'])
```

```python
name.drop(['Insulin_x', 'Insulin_y'], axis=1, inplace=True)
```

```python
name
```

```python
from sklearn.preprocessing import MinMaxScaler

columns = ["BMI", "Insuline", 'Glucose', 'BloodPressure', 'SkinThickness', 'Age']
for i in columns:
    scaler_ = MinMaxScaler()
    print(i)
    name[i] = scaler_.fit_transform(np.array(name[i]).reshape(-1, 1))
```

```python
name
```

```python
ohe_cols = ['Pregnancies']
for i in ohe_cols:
    dummy = pd.get_dummies(name[i], prefix = i).astype(int)
    mask = str(name[i].value_counts().index[0])
    dummy.drop(i + '_' + mask, inplace=True, axis=1)
    finaltable = pd.concat([name,dummy], axis=1)
    finaltable.drop(i, inplace=True, axis=1)
```

```python
finaltable
```

```python
X = finaltable.drop('Outcome',axis=1)
y = finaltable['Outcome']

from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.3, random_state = 21 )
```

```python
X_train
```

```python
y_train
```

```python
from sklearn.linear_model import LogisticRegression
lr1 = LogisticRegression(random_state=33)
lr1.fit(X_train,y_train)
y_pred = lr1.predict(X_test)
confusion_matrix = pd.crosstab(y_test, y_pred, rownames=['Actual'], colnames=['Predicted'])
confusion_matrix
```

```python
recall_score = confusion_matrix[0][0]/(confusion_matrix[0][0]+confusion_matrix[0][1])
recall_score
```

```python
precision_score = confusion_matrix[0][0]/(confusion_matrix[0][0]+confusion_matrix[1][0])
precision_score
```

```python
f1_score = 2*recall_score*precision_score/(recall_score+precision_score)
f1_score
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

fig, ax = plt.subplots()
sns.heatmap(confusion_matrix, annot=True, fmt='g', ax=ax)
ax.set_xlabel('Predicted labels')
ax.set_ylabel('True labels')
ax.set_title('Diabetes outcome')
ax.xaxis.set_ticklabels(["No Diabetes", 'Diabetes'])
ax.yaxis.set_ticklabels(["No Diabetes", 'Diabetes'])
#plt.savefig(save_to + "/Rfc_test.svg")
plt.show()
```
